```python
def get_response(number):

    url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

    querystring = {"id":number,"season":"2020"}

    headers = {
        "X-RapidAPI-Key": "bddfdf70b8mshbb1ca9870c018e7p119c4ejsn7cd1c7ceb114",
        "X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
    }
    
    response = requests.request("GET", url, headers=headers, params=querystring)
    if response:
        return response
    return None
```

```python
responses=[]
_min=140
_max=_min+20
for i in range(_min,_max):
    r=get_response(str(i))
    if r:
        responses.append(r)
```

```python
valid_responses=[]
for i in responses:
    if i:
        try:
#             print (i.json()['response'][0]['player']['firstname'])
            if len(i.json()['response'])>0:
                valid_responses.append(i)
        except:
#             print(i.json())
            continue
# for i in valid_responses:
#     print(i.json()['response'][0]['player'])
valid_responses
```

```python
#מחר לטפל בתא הזה
flag=True
cnt=0
while flag:
    try:
        temp=valid_responses[cnt].json()['response'][cnt].keys()
        flag = False
    except:
        cnt+=1

RES={}
for key in temp:
    RES[key]=[]
res=RES.copy()
```

```python
for i in valid_responses:
    player=i.json()['response']
    
    for row in player:
        for v in row:
            if v=='game':
                res['game'].append(row[v]['id'])
            elif v=='player':
                res[v].append(row[v]['firstname']+"_"+row[v]['lastname'])
            elif v=='team':
                res[v].append(row[v]['name'])
            else: 
                res[v].append(row[v])
    
```

```python

df=pd.DataFrame(res)
df
```

```python
df.to_csv(f'{str(_min)}-{str(_max)}.csv')
```
